```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Read the City and Ride Data
city_data = pd.read_csv("data/city_data.csv")
ride_data = pd.read_csv("data/ride_data.csv")
```

```python
# Combine the data into a single dataset and display
combined_data = pd.merge(city_data, ride_data, on="city")
combined_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

Urban = combined_data[(combined_data.type=="Urban")]
Suburban = combined_data[(combined_data.type=="Suburban")]
Rural = combined_data[(combined_data.type=="Rural")]

Urban_rides_per_city =Urban.groupby([combined_data.city]).count()["ride_id"]
Urban_average_fare = Urban.groupby([combined_data.city]).mean()["fare"]

Suburban_rides_per_city =Suburban.groupby([combined_data.city]).count()["ride_id"]
Suburban_average_fare = Suburban.groupby([combined_data.city]).mean()["fare"]

Rural_rides_per_city =Rural.groupby([combined_data.city]).count()["ride_id"]
Rural_average_fare = Rural.groupby([combined_data.city]).mean()["fare"]

# Build the scatter plots for each city types

plt.scatter(Urban_rides_per_city, Urban_average_fare, label = "Urban", s=Urban_rides_per_city*15, marker="o", facecolors=["lightcoral"], edgecolors="black")
plt.scatter(Suburban_rides_per_city, Suburban_average_fare, label = "Suburban", s=Suburban_rides_per_city*8, marker="o", facecolors=["lightskyblue"], edgecolors="black")
plt.scatter(Rural_rides_per_city, Rural_average_fare, label = "Rural", s=Urban_rides_per_city*5, marker="o", facecolors=["gold"], edgecolors="black")

# Incorporate the other graph properties
plt.Xlim = (0,40)
plt.ylim(15,50)
    
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides per city")
plt.ylabel("Average Fare ($)")

# Create a legend

plt.legend(title = "City Types")

# Incorporate a text label regarding circle size
plt.annotate("Note: \nCircle size correlates with driver count per city.", xy=(30, 30), xycoords='data',xytext=(42.5, 35),)

# Show plot
plt.grid(c="grey")
```

## Total Fares by City Type

```python
# Calculate Type Percents

total_fares_by_city = combined_data.groupby(["type"])[["fare"]].sum()
```

```python
# Build Pie Chart

plt.pie(total_fares_by_city,
         explode = [0,0,0.1],
         labels = ["Rural","Suburban", "Urban"],
         colors = ["gold","lightskyblue", "lightcoral"],
         autopct = '%.1f%%',
         pctdistance = 0.5,
         shadow = True,
         startangle = 120)

plt.title("% of Total Fares by CIty Type")
plt.axis("equal")          
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_Rides_by_city = combined_data.groupby(["type"])[["ride_id"]].count()

```

```python
# Build Pie Chart

plt.pie(total_Rides_by_city,
         explode = [0,0,0.1],
         labels = ["Rural","Suburban", "Urban"],
         colors = ["gold","lightskyblue", "lightcoral"],
         autopct = '%.1f%%',
         pctdistance = 0.5,
         shadow = True,
         startangle = 120)

plt.title("% of Total Rides by CIty Type")
plt.axis("equal")          
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

total_drivers_by_city = city_data.groupby(["type"])[["driver_count"]].sum()
```

```python
# Build Pie Charts

plt.pie(total_drivers_by_city,
         explode = [0,0,0.1],
         labels = ["Rural","Suburban", "Urban"],
         colors = ["gold","lightskyblue", "lightcoral"],
         autopct = '%.1f%%',
         pctdistance = 0.5,
         shadow = True,
         startangle = 120)

plt.title("% of Total Drivers by CIty Type")
plt.axis("equal")          
plt.show()
```
